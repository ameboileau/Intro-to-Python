```python
# Importing packages...

import numpy as np
import matplotlib.pyplot as plt
```

```python
def group_mean(r, n):
    '''
    Divide r into groups of n, and taking the mean of each group
    r: array of random variables (numpy array)
    n: number of elements per group (int)
    '''
    means = []
    for i in range(0, r.size, n):
        group = []
        for num in range(n):
            group.append(r[int(i+num)])
        means.append(np.mean(group))
        
    return np.asarray(means)


x = np.arange(0,1,0.001)
r = np.random.uniform(0,1,size=1000000)

plt.hist(r, bins=50, density=True)
plt.xlabel("Random variable value")
plt.ylabel("PDF")
plt.show()
```
